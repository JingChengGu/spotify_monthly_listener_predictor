### Data Cleaning and Preperation

```python
spotify_artist_info.head()
```

```python
spotify_artist_info.columns
```

```python
# remove columns
# we dont need ids
# we dont need feat_track_ids
# we dont need geners because there is too much noise in the categories.
# we dont need last release since there is cluster on 20 since any songs above 20 is counted within
# we dont need num_releases
spotify_artist_info = spotify_artist_info[['dates', 
                                           #'ids', 
                                           'names', 'monthly_listeners', 'popularity', 'followers','genres', 
                                           'first_release', 'last_release'
                                           #, 'num_releases', 'num_tracks'
                                           , 'playlists_found']].copy()
                                           #, 'feat_track_ids'
```

```python
spotify_artist_info.shape
```

```python
# rename columns
spotify_artist_info = spotify_artist_info.rename(columns = {
    'first_release' : 'first_release_year',
    'last_release' : 'last_release_year',
})
```

```python
spotify_artist_info.isna().sum()
```

```python
# discovering genres
spotify_artist_info['genres'].value_counts()
```

```python
spotify_artist_info[spotify_artist_info['monthly_listeners'].isna()].head()
```

```python
spotify_artist_info = spotify_artist_info[spotify_artist_info["monthly_listeners"].notna()]
```

```python
spotify_artist_info.isna().sum()
```

```python
spotify_artist_info.shape
```

```python
# remove first release year that is -1 and 0
spotify_artist_info = spotify_artist_info[(spotify_artist_info['first_release_year'] != -1) & (spotify_artist_info['first_release_year'] != 0)]
```

```python
spotify_artist_info.loc[spotify_artist_info.duplicated()]
```

```python
spotify_artist_info.loc[spotify_artist_info.duplicated('names')]
```

```python
unique_artists = spotify_artist_info[~spotify_artist_info.duplicated(subset = ['names'])]
```

### 3. Feature Understanding
- Univariate Analysis
- Plotting Feature Distributions

```python
spotify_artist_info.head()
```

#### Release Year Analysis

```python
release_year_plot = spotify_artist_info['first_release_year'].value_counts().head(10) \
        .plot(kind = 'bar', title = 'Top Years Artists First Release')

release_year_plot.set_xlabel('Year Artist First Released')
release_year_plot.set_ylabel('Count')
```

```python
release_year_hist = spotify_artist_info['first_release_year'].plot(kind = 'hist', bins = 20 \
                                                ,title = 'Distribution of Release Year')
release_year_hist.set_xlabel('Year')
release_year_hist.set_ylabel('Count')
```

#### Log Scale Followers Analysis

```python
spotify_artist_info['followers']
```

```python
log_followers = np.log10(spotify_artist_info['followers'].replace(0, np.nan))
```

```python
log_followers_hist = log_followers.plot(kind = 'hist', bins = 8, title = "Distribution of Followers (Log Scale)")
log_followers_hist.set_xlabel('Log10 of Follower Count')
log_followers_hist.set_ylabel('Number of Artsts')
```

```python
log_followers_kde = log_followers.plot(kind = 'kde', title = "Density of Followers (Log Scale)")
log_followers_kde.set_xlabel('Log10 of Follower Count')
log_followers_kde.set_ylabel('Number of Artsts')
```

#### Popularity Analysis

```python
spotify_artist_info['popularity'].value_counts()
```

```python
popularity_hist = spotify_artist_info['popularity'].plot(kind = 'hist', bins = 10, title= 'Popularity Distribution')
popularity_hist.set_xlabel('Popularity Score')
popularity_hist.set_ylabel('Count')
```

```python
popularity_kde = spotify_artist_info['popularity'].plot(kind = 'kde', title= 'Popularity Density')
popularity_kde.set_xlabel('Popularity Score')
popularity_kde.set_ylabel('Density')
```

#### Monthly Listeners Analysis (log scaled)

```python
log_monthly_listeners = np.log10(spotify_artist_info['monthly_listeners'].replace(0, np.nan)).dropna().round().astype(int)
```

```python
log_monthly_listeners_hist = log_monthly_listeners.plot(kind = 'hist', bins = 8, title = 'Distribution of Monthly Listeners (Log Scaled)')
log_monthly_listeners_hist.set_xlabel('Monthly Listeners (Log Scaled)')
log_monthly_listeners_hist.set_ylabel('Count')
```

```python
log_monthly_listeners_kde = log_monthly_listeners.plot(kind = 'kde', title = 'Density of Monthly Listeners (Log Scaled)')
log_monthly_listeners_kde.set_xlabel('Monthly Listeners (Log Scaled)')
log_monthly_listeners_kde.set_ylabel('Density')
```
